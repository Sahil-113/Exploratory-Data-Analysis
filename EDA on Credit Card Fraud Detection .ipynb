# Visual Analysis

```python
class_percentages=(credit['Class']).value_counts()/len(credit)*100
#Create the Pie chart
plt.figure(figsize=(5,5))
plt.pie(class_percentages,labels=class_percentages.index,autopct='%.1f%%',startangle=30)
plt.title("Class Percentage")
plt.axis('equal')
plt.savefig("class_percentage.png",bbox_inches='tight')
plt.show()
```

Only 0.2% of the dataset consists fraudlent Transactions


 Blue line of next graph represent the normal transaction and red line represents the fraudlent transaction

```python
sns.catplot(x="Class",kind="count",data=credit,height=5)
plt.title("Class Count",fontsize=18)
plt.xlabel("Fraud",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

Class is Highly imbalenced. generally, in fraud detection, class imbalence is very common.

```python
timedelta=pd.to_timedelta(credit['Time'],unit='s')
credit["Time_min"]=(timedelta.dt.components.minutes).astype(int)
credit['Time_hour']=(timedelta.dt.components.hours).astype(int)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
plt.figure(figsize=(12,5))
sns.distplot(credit[credit['Class']==0]['Time_hour'],color='b')
sns.distplot(credit[credit['Class']==1]['Time_hour'],color='r')
plt.title("Fraud x Normal Transaction By Hours",fontsize=17)
plt.xlim([-1,25])
plt.show()
```

```python
#By Minute
plt.figure(figsize=(12,5))
sns.distplot(credit[credit['Class']==0]['Time_min'],color='b')
sns.distplot(credit[credit['Class']==1]['Time_min'],color='r')
plt.title("Fraud x Normal Transaction By Hours",fontsize=17)
plt.xlim([-1,61])
plt.show()
```

```python
# Descriptive Statistics for fraud and non fraud transactions
fraud =credit[credit['Class']==1]
normal=credit[credit['Class']==0]
print("Fraud Transaction Statistics")
print(fraud["Amount"].describe())
print("\nNormal Transaction Statistics")
print(normal['Amount'].describe())
```

```python
plt.figure(figsize=(14,6))
# exploring the amount by class and the distribution of amount transactions
plt.subplot(121)
ax=sns.boxplot(x='Class',y='Amount',data=credit)
ax.set_title("Class x Amount",fontsize=20)
ax.set_xlabel("Fraud",fontsize=16)
ax.set_ylabel("Amount(US)",fontsize=16)

plt.subplot(122)
ax=sns.boxplot(x='Class',y=np.log1p(credit['Amount']),data=credit)
ax.set_title("Class x Amount",fontsize=20)
ax.set_xlabel("Fraud",fontsize=16)
ax.set_ylabel("Amount(Log)",fontsize=16)
plt.subplots_adjust(hspace=0.6,top=0.8)
plt.show()
```

From the figure we can see that 0(non fraudulent transection) tends to have higher transaction amount than the
1 (fraudulent transection)

the second subplot show the same distribution but with the logrithemic transformation applied to the amount values.

Using the logrithemic scale provide clearer visualisation of distribution, especially when there are extreme values or 
outliers in the transaction amount

```python
# Looking the amount and time distribution of fraud transaction
ax=sns.lmplot(y='Amount',x='Time_min',fit_reg=False,aspect=1.8,data=credit,hue='Class')
plt.title("Amounts by Minutes of Frauds and Normal Transaction",fontsize=16)
plt.show()
```

There are no clearer patterns or trends in the time of day when transaction occur

It appear that the transaction time follow a unique distribution throughout recorded time period

```python
# Correlation Matrix
plt.figure(figsize=(12,8))
corr=credit.corr()
sns.heatmap(corr,cmap='PRGn',annot=False,fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

The correlation Matrix reveals weak corelation between most feature, indicating low linear dependence between them.
